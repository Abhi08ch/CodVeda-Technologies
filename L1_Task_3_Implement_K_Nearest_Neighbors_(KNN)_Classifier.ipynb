# **Task 3: Implement K-Nearest Neighbors (KNN) Classifier**

```python
# 1. Import all the libraries required
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
# 2. Load and prepare the dataset
df = pd.read_csv("/content/1) iris.csv")

if 'ID' in df.columns:
  df = df.drop('ID', axis=1)

# Separate features and target
x = df.iloc[:, :-1] # first 4 columns: features
y = df.iloc[:, -1]  # last column: species
```

```python
# 3.  Split Data into Training and Test Sets
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=42, stratify=y)
```

```python
# 4. Feature Scaling
scalar = StandardScaler()
x_train_scaled = scalar.fit_transform(x_train)
x_test_scaled = scalar.transform(x_test)
```

```python
# 5. Train the KNN Classifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train_scaled, y_train)
```

```python
# 6. Make predictions
y_pred = knn.predict(x_test_scaled)
```

```python
# 7. Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy_score: {accuracy:.2f}")

print("Confusion Matrix: ")
print(confusion_matrix(y_test, y_pred))

print("Classififcation report: ")
print(classification_report(y_test, y_pred))
```

```python
# 8. Vizulaizinng the confusion matrix
ConfusionMatrixDisplay.from_estimator(knn, x_test_scaled, y_test, cmap='Blues')
plt.title("KNN Confusion Matrix (Iris Dataset)")
plt.show()
```
